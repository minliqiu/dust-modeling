```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# add new index

para_new = np.array([[1.  ,  0.3 , 10.  ,  0.01, 10.  ,  0.74,  0.26,  0.  ,  0.]])
paralabels = ["m_Star/m_Sun", "m_Planet/m_J", "a_p/R_Star", "beta", "inc_d", "f_sub", "f_pl", "f_ej", "f_lf"]
df_para_new = pd.DataFrame(para_new, columns = paralabels)
```

```python
df_para = pd.concat([df_para, df_para_new], axis=0)
```

```python
df_para = df_para.sort_values(by=['m_Star/m_Sun', 'm_Planet/m_J', 'beta'], ascending = False)
df_para = df_para.reset_index(drop = True)
# df_para.to_csv('./result.csv', header=True, index=False)

df_para
```

```python
# smaller star & medium planet
df_beta05_s = df_para[(df_para['m_Planet/m_J']==1) & (df_para['beta']==0.5)]
df_beta01_s = df_para[(df_para['m_Planet/m_J']==1) & (df_para['beta']==0.1)]
df_beta001_s = df_para[(df_para['m_Planet/m_J']==1) & (df_para['beta']==0.01)]
```

```python
fig, ax = plt.subplots(3, 1, figsize=(8,14))

ax[0].text(0.64, 1.5, 'Smaller Star', fontsize=22, ha='center')
ax[0].text(0.64, 1.37, 'm_p/m_J=1, a_p/R_Star=10, inc_d = 10', fontsize=17, ha='center')
ax[0].set_title('beta = 0.01')
ax[0].set_ylabel("fraction", fontsize=20)
ax[0].set_xlim([0, 1.25])
ax[0].set_ylim([0, 1.2])
ax[0].bar(df_beta001_s["m_Star/m_Sun"], df_beta001_s["f_pl"]+df_beta001_s["f_sub"]+df_beta001_s["f_ej"]+df_beta001_s["f_lf"], width = 0.07, color = 'gold', label = 'f_lf')
ax[0].bar(df_beta001_s["m_Star/m_Sun"], df_beta001_s["f_pl"]+df_beta001_s["f_sub"]+df_beta001_s["f_ej"], width = 0.07, color = 'mediumaquamarine', label = 'f_ej')
ax[0].bar(df_beta001_s["m_Star/m_Sun"], df_beta001_s["f_pl"]+df_beta001_s["f_sub"], width = 0.07, color = 'salmon', label = 'f_sub')
ax[0].bar(df_beta001_s["m_Star/m_Sun"], df_beta001_s["f_pl"], width = 0.07, color = 'cornflowerblue', label = 'f_pl')
ax[0].legend(loc='upper right')

ax[1].set_title('beta = 0.1')
ax[1].set_ylabel("fraction", fontsize=20)
ax[1].set_xlim([0, 1.25])
ax[1].set_ylim([0, 1.2])
ax[1].bar(df_beta01_s["m_Star/m_Sun"], df_beta01_s["f_pl"]+df_beta01_s["f_sub"]+df_beta01_s["f_ej"]+df_beta01_s["f_lf"], width = 0.07, color = 'gold', label = 'f_lf')
ax[1].bar(df_beta01_s["m_Star/m_Sun"], df_beta01_s["f_pl"]+df_beta01_s["f_sub"]+df_beta01_s["f_ej"], width = 0.07, color = 'mediumaquamarine', label = 'f_ej')
ax[1].bar(df_beta01_s["m_Star/m_Sun"], df_beta01_s["f_pl"]+df_beta01_s["f_sub"], width = 0.07, color = 'salmon', label = 'f_sub')
ax[1].bar(df_beta01_s["m_Star/m_Sun"], df_beta01_s["f_pl"], width = 0.07, color = 'cornflowerblue', label = 'f_pl')
ax[1].legend(loc='upper right')

ax[2].set_title('beta = 0.5')
ax[2].set_xlabel("m_Star/m_Sun", fontsize=20)
ax[2].set_ylabel("fraction", fontsize=20)
ax[2].set_xlim([0, 1.25])
ax[2].set_ylim([0, 1.2])
ax[2].bar(df_beta05_s["m_Star/m_Sun"], df_beta05_s["f_pl"]+df_beta05_s["f_sub"]+df_beta05_s["f_ej"]+df_beta05_s["f_lf"], width = 0.07, color = 'gold', label = 'f_lf')
ax[2].bar(df_beta05_s["m_Star/m_Sun"], df_beta05_s["f_pl"]+df_beta05_s["f_sub"]+df_beta05_s["f_ej"], width = 0.07, color = 'mediumaquamarine', label = 'f_ej')
ax[2].bar(df_beta05_s["m_Star/m_Sun"], df_beta05_s["f_pl"]+df_beta05_s["f_sub"], width = 0.07, color = 'salmon', label = 'f_sub')
ax[2].bar(df_beta05_s["m_Star/m_Sun"], df_beta05_s["f_pl"], width = 0.07, color = 'cornflowerblue', label = 'f_pl')
ax[2].legend(loc='upper right')


plt.show()
```
